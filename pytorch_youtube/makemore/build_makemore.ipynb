### Broadcasting

Can we divide 27x27 array with 27x1 array?

&rarr; Broadcasting!

Two tensors are "broadcastable" if the following rules hold:
> Each tensor has at least one dimension

> When iterating over the dimension sizes, started at the trailing dimension, the dimension sizes must either be equal, one of them is 1, or one of them does not exist.

What does it actually do?
Copies the column vector 27 times to make it also 27x27 size array, then do the element wise division.

''' P.sum(1).shape ''' 

You would expect this to work becasue after we align to the right side, we do not want to broadcast this way. That is why we ned keepdim=True. If the dimension is reduced. it becomes size 27 instead of 1x27. Then during broadcasting, the right side is aligned as it changes to 1x27 which is a row vector. When torch replicates its dimension, it will replicated vertically instead of horizontally. We are normalizing columnds instead of rows. 

```python
P.sum(1).shape
```

```python
# Model training step
P = N.float()
P /= P.sum(1, keepdim=True)  # dont change the dimension!!
```

```python
P[0].sum() # normalized
```

```python
P[:, 0].sum() # not normalized 
```

```python
# Sampling step

g = torch.Generator().manual_seed(2147483647)

for i in range(5):
    out = []
    ix = 0 # start token
    while True:
        p = P[ix]
        ix = torch.multinomial(p, num_samples=1, replacement=True, generator=g).item()
        out.append(itos[ix])
        # print(itos[ix])
        if ix == 0: # end token
            break
    print(''.join(out))
```

### Loss Function 
We want to summarize/evaluate how good a model is with a single number 

```python
log_likelihood = 0.0
n = 0
for w in words[:3]:
    chs = ['.'] + list(w) + ['.']
    for ch1, ch2 in zip(chs, chs[1:]):
        ix1 = stoi[ch1]
        ix2 = stoi[ch2]
        prob = P[ix1, ix2]
        logprob = torch.log(prob)
        log_likelihood += logprob
        n += 1
        print(f'{ch1}{ch2}: {prob:.4f} {logprob:.4f}')

print(f'{log_likelihood=}')

nll = -log_likelihood
print(f'{nll=}')
print(f'{nll/n}')
```
